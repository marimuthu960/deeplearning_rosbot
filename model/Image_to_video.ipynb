```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import random
import time
import tensorflow as tf
import h5py
from keras.models import load_model
import sys
import glob
from natsort import natsorted
from moviepy.editor import *
import csv
%matplotlib inline
```

```python
#checking the dataset
count = 1
which = 2
how_many = 500
how_many_count = 0
modified = []
for root, directories, filenames in os.walk('drive_data/training_data'):
    for directory in directories:
        if count == which:
            print(directory)
            file_path = root+"/"+directory+"/"
            data_file_path  = file_path+'data.csv'
            f = open(data_file_path, 'r')
            reader = csv.reader(f)
            for row in reader:
                if count != 0 and '.jpg' in row[0]:
                    image_path = file_path+row[0]
                    steering = row[1]
                    throttle = row[2]
                    last_image1 = cv2.imread(image_path,cv2.IMREAD_COLOR)
                    last_image1 = cv2.resize(last_image1, (320, 240), interpolation=cv2.INTER_AREA)
                    cv2.putText(last_image1, 'Image: {}'.format(str(row[0])), (10, 40), cv2.FONT_HERSHEY_SIMPLEX, .5, (255, 255, 255), 1)
                    cv2.putText(last_image1, 'Steering: {}'.format(str(steering)), (10, 60), cv2.FONT_HERSHEY_SIMPLEX, .5, (255, 255, 255), 1)
                    modified.append(last_image1)
                if how_many_count > how_many: 
                    break
                else:
                    how_many_count += 1
        count +=1
fps = 30
clips = [ImageClip(m).set_duration(1) for m in modified]
concat_clip = concatenate_videoclips(clips, method="compose")
#concat_clip = concat_clip.speedx(factor=10)
concat_clip.write_videofile("test_video.mp4", fps=fps)
```

```python
def linear_unbin1(arr):
    b = np.argmax(arr)
    a = b *(2/14) - 1
    return a
```

```python
base_dir = os.path.realpath("./test_model_images1")
print(base_dir)

gif_name = 'pic'
fps = 30

file_list = glob.glob(base_dir+'/*.jpg')  # Get all the pngs in the current directory
file_list_sorted = natsorted(file_list,reverse=False)  # Sort the images
model = load_model("model.h5")
modifed = []
for m in file_list_sorted:
    images = []
    last_image1 = cv2.imread(m,cv2.IMREAD_COLOR)
    last_image1 = cv2.resize(last_image1, (160, 120), interpolation=cv2.INTER_AREA)
    images.append(np.asarray(last_image1))
    value = model.predict(np.array(images), batch_size=1)
    value = linear_unbin1(value)
    cv2.putText(last_image1, 'Steering: {}'.format(str(value)), (10, 60), cv2.FONT_HERSHEY_SIMPLEX, .5, (255, 255, 255), 1)
    modifed.append(last_image1)
    
clips = [ImageClip(m).set_duration(1)
         for m in modifed]

concat_clip = concatenate_videoclips(clips, method="compose")
concat_clip.write_videofile("test_video.mp4", fps=fps)
```
